### Initializing mediapipe pose class.

To start, we must first initialize the pose class using the **`mp.solutions.pose`** syntax. After that, we will proceed by calling the setup function **`mp.solutions.pose.Pose()`** with the given arguments:

When `static_image_mode` is set to `False`, the detector is only called when necessary - such as in the first frame or when the tracker cannot follow the target. If set to `True`, the person detector will be called for each input image. It's recommended to set this value to `True` when dealing with many individual images, not videos. The default setting is `False`.

The `min_detection_confidence` is the minimum confidence level required (ranging from 0.0 to 1.0) for the person-detection model to make a correct prediction. By default, this value is set to 0.5, indicating that any predictions with a confidence level of 50% or higher will be considered as valid detections.


The `min_tracking_confidence` parameter determines the minimum level of confidence (ranging from 0.0 to 1.0) needed for the landmark-tracking model to deem the tracked pose landmarks as valid. When the confidence falls below this threshold, the detector is called again in the subsequent frame or image. Raising this value enhances reliability but also adds to the processing delay. By default, `min_tracking_confidence` is set to 0.5.


- The `model_complexity` parameter determines how intricate the pose landmark model is. Options include `0`, `1`, or `2`, with higher values offering more accurate results but also increasing latency. The default value is `1`. 

- The `smooth_landmarks` parameter is a boolean that, when set to `True`, filters pose landmarks across frames to reduce noise. This feature only works when `static_image_mode` is set to `False`. The default value is `True`. 

- Additionally, we will instantiate the `mp.solutions.drawing_utils` class to visually represent the detected landmarks. Alternatively, OpenCV can be used for visualization instead of this class.

```python
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=True, min_detection_confidence=0.5, model_complexity=1)
mp_drawing = mp.solutions.drawing_utils 
```

## **<font style="color:rgb(134,19,348)">Perform Pose Detection</font>**

**Next, we will feed the image into the pose detection machine learning pipeline using the function **`mp.solutions.pose.Pose().process()`**. 

The pipeline, however, requires the input images to be in **`RGB`** color format. So, we will first need to convert the sample image from **`BGR`** to **`RGB`** format using the function [**`cv2.cvtColor()`**](https://docs.opencv.org/3.4/d8/d01/group__imgproc__color__conversions.html#ga397ae87e1288a81d2363b61574eb8cab) since OpenCV reads images in **`BGR`** format (as opposed to **`RGB`). 

Upon completion of pose detection, we will receive a list of thirty-three landmarks that indicate the body joint locations of the prominent person

After analyzing the pose, we will receive a list of thirty-three points showing where the person's body joints are located in the picture. Each point includes: 

* **`x`** - This is the point's horizontal position, scaled between 0.0 and 1.0 based on the image's width. 

* **`y`** - This is the point's vertical position, scaled between 0.0 and 1.0 based on the image's height. 

* **`z`** - This is the point's depth, scaled similarly to **`x`**. It indicates how far the point is from the camera, with the hips as the reference point. 

 * **`visibility`** - This is a value between 0.0 and 1.0, showing the likelihood of the point being visible in the image without obstruction.

```python
def detectPose(image, pose, display=True):
    '''
    This function performs pose detection on an image.
    Args:
        image: The input image with a prominent person whose pose landmarks needs to be detected.
        pose: The pose setup function required to perform the pose detection.
        display: A boolean value that is if set to true the function displays the original input image, the resultant image, 
                 and the pose landmarks in 3D plot and returns nothing.
    Returns:
        output_image: The input image with the detected pose landmarks drawn.
        landmarks: A list of detected landmarks converted into their original scale.
    '''
    
    # Create a copy of the input image.
    output_image = image.copy()
    
    # Convert the image from BGR into RGB format.
    imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Perform the Pose Detection.
    results = pose.process(imageRGB)
    
    # Retrieve the height and width of the input image.
    height, width, _ = image.shape
    
    # Initialize a list to store the detected landmarks.
    landmarks = []
    
    # Check if any landmarks are detected.
    if results.pose_landmarks:
    
        # Draw Pose landmarks on the output image.
        mp_drawing.draw_landmarks(image=output_image, landmark_list=results.pose_landmarks,
                                  connections=mp_pose.POSE_CONNECTIONS)
        
        # Iterate over the detected landmarks.
        for landmark in results.pose_landmarks.landmark:
            
            # Append the landmark into the list.
            landmarks.append((int(landmark.x * width), int(landmark.y * height),
                                  (landmark.z * width)))
    
    # Check if the original input image and the resultant image are specified to be displayed.
    if display:
    
        # Display the original input image and the resultant image.
        plt.figure(figsize=[22,22])
        plt.subplot(121);plt.imshow(image[:,:,::-1]);plt.title("Original Image");plt.axis('off');
        plt.subplot(122);plt.imshow(output_image[:,:,::-1]);plt.title("Output Image");plt.axis('off');
        
        # Also Plot the Pose landmarks in 3D.
        mp_drawing.plot_landmarks(results.pose_world_landmarks, mp_pose.POSE_CONNECTIONS)
        
    # Otherwise
    else:
        
        # Return the output image and the found landmarks.
        return output_image, landmarks
```

## "**<font style="color:rgb(134,19,348)">Real-Time Pose Detection on Webcam Feed/Video</font>**"" 

The performance on the images was impressive. Next, we aim to test the functionality on a live webcam feed and a video. Depending on whether we wish to conduct pose detection on a video saved on the computer or on the webcam feed, adjust the initialization code of the VideoCapture object by commenting and uncommenting as needed."

```python
# Setup Pose function for video.
pose_video = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, model_complexity=1)

# Initialize the VideoCapture object to read from an MP4 video stored in the disk.
video = cv2.VideoCapture('media/playbook/plank.mp4')

# Create named window for resizing purposes
cv2.namedWindow('Pose Detection', cv2.WINDOW_NORMAL)

# Set video size
video.set(3, 1280)
video.set(4, 960)

# Initialize a variable to store the time of the previous frame.
time1 = 0

# Iterate until the video is accessed successfully.
while video.isOpened():
    
    # Read a frame.
    ok, frame = video.read()
    
    # Check if frame is not read properly.
    if not ok:
        
        # Break the loop.
        break
    
    # Flip the frame horizontally for natural (selfie-view) visualization.
    frame = cv2.flip(frame, 1)
    
    # Get the width and height of the frame
    frame_height, frame_width, _ =  frame.shape
    
    # Resize the frame while keeping the aspect ratio.
    frame = cv2.resize(frame, (int(frame_width * (640 / frame_height)), 640))
    
    # Perform Pose landmark detection.
    frame, _ = detectPose(frame, pose_video, display=False)
    
    # Set the time for this frame to the current time.
    time2 = time()
    
    # Check if the difference between the previous and this frame time > 0 to avoid division by zero.
    if (time2 - time1) > 0:
    
        # Calculate the number of frames per second.
        frames_per_second = 1.0 / (time2 - time1)
        
        # Write the calculated number of frames per second on the frame. 
        cv2.putText(frame, 'FPS: {}'.format(int(frames_per_second)), (10, 30),cv2.FONT_HERSHEY_PLAIN, 2, (0, 255, 0), 3)
    
    # Update the previous frame time to this frame time.
    # As this frame will become previous frame in the next iteration.
    time1 = time2
    
    # Display the frame.
    cv2.imshow('Pose Detection', frame)
    
    # Wait until a key is pressed.
    # Retreive the ASCII code of the key pressed
    k = cv2.waitKey(1) & 0xFF
    
    # Check if 'ESC' is pressed.
    if(k == 27):
        
        # Break the loop.
        break

# Release the VideoCapture object.
video.release()

# Close the windows.
cv2.destroyAllWindows()
```

## **<font style="color:rgb(134,19,348)">Pose Classification with Angle Heuristics</font>**

We performed the pose detection, Since we now have about 36 features to work with, it will greatly reduce the performamce of our model, thus we came up with a heuristic to reduce the 36 features to 6 features by calculating the angles between joints and depending on these angles we will recognise the yoga positions in an image or videos. 


```python
def calculateAngle(landmark1, landmark2, landmark3):
    '''
    This function calculates angle between three different landmarks.
    Args:
        landmark1: The first landmark containing the x,y and z coordinates.
        landmark2: The second landmark containing the x,y and z coordinates.
        landmark3: The third landmark containing the x,y and z coordinates.
    Returns:
        angle: The calculated angle between the three landmarks.

    '''

    # Get the required landmarks coordinates.
    x1, y1, _ = landmark1
    x2, y2, _ = landmark2
    x3, y3, _ = landmark3

    # Calculate the angle between the three points
    angle = math.degrees(math.atan2(y3 - y2, x3 - x2) - math.atan2(y1 - y2, x1 - x2))
    
    # Check if the angle is less than zero.
    if angle < 0:

        # Add 360 to the found angle.
        angle += 360
    
    # Return the calculated angle.
    return angle
```

### **<font style="color:rgb(134,19,348)">Create a Function to Perform by loading the Pose Classification Model</font>**

Now we will use a function that will be capable of classifying different yoga poses using the Model that's been trained on calculated angles of various joints.

```python
def PoseBuilder(landmarks, output_image, display=False):
    '''
    This function classifies yoga poses depending upon the angles of various body joints.
    Args:
        landmarks: A list of detected landmarks of the person whose pose needs to be classified.
        output_image: A image of the person with the detected pose landmarks drawn.
        display: A boolean value that is if set to true the function displays the resultant image with the pose label
        written on it and returns nothing.
    Returns:
        output_image: The image with the detected pose landmarks drawn and pose label written.
        label: The classified pose label of the person in the output_image.

    '''
    # Calculate the required angles.
    #----------------------------------------------------------------------------------------------------------------    
    # Get the angle between the left shoulder, elbow and wrist points.
    left_elbow_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value],
                                      landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value],
                                      landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value])

    # Get the angle between the right shoulder, elbow and wrist points.
    right_elbow_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value],
                                       landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value],
                                       landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value])

    # Get the angle between the left elbow, shoulder and hip points.
    left_shoulder_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value],
                                         landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value],
                                         landmarks[mp_pose.PoseLandmark.LEFT_HIP.value])

    # Get the angle between the right hip, shoulder and elbow points.
    right_shoulder_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value],
                                          landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value],
                                          landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value])

    # Get the angle between the left hip, knee and ankle points.
    left_knee_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value],
                                     landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value],
                                     landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value])

    # Get the angle between the right hip, knee and ankle points
    right_knee_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value],
                                      landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value],
                                      landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value])

    # Get the angle between the left hip, knee and ankle points.
    left_hip_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value],
                                     landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value],
                                     landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value])

    # Get the angle between the right hip, knee and ankle points
    right_hip_angle = calculateAngle(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value],
                                      landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value],
                                      landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value])
    
    return  left_elbow_angle, right_elbow_angle, left_shoulder_angle, right_shoulder_angle, left_knee_angle, right_knee_angle, left_hip_angle, right_hip_angle
```

```python
# Load the stacking classifier model
with open('stacking_classifier_model.pkl', 'rb') as f:
    loaded_model = pickle.load(f)

# Function to classify pose using the loaded model
def classifyPose(landmarks, output_image, display=False, threshold=0.8):
    '''
    Classify yoga poses using the loaded model and annotate the image with the classification.
    Args:
        landmarks: Detected landmarks of the person whose pose needs to be classified.
        output_image: Image of the person with the detected pose landmarks drawn.
        display: Boolean flag to indicate whether to display the results.
        threshold: Minimum confidence/probability threshold to classify a pose confidently.
    Returns:
        output_image: Image with the classified pose label written.
        label: Classified pose label.
    '''
    # Calculate features from landmarks
    # Calculate features from landmarks
    left_elbow_angle, right_elbow_angle, left_shoulder_angle, right_shoulder_angle, left_knee_angle, right_knee_angle, left_hip_angle, right_hip_angle = PoseBuilder(landmarks, output_image)
    
    # Prepare features for prediction
    features_df = pd.DataFrame({'left_elbow_angle': [left_elbow_angle], 'right_elbow_angle': [right_elbow_angle], 'left_shoulder_angle': [left_shoulder_angle], 'right_shoulder_angle': [right_shoulder_angle], 'left_knee_angle': [left_knee_angle], 'right_knee_angle': [right_knee_angle], 'left_hip_angle': [left_hip_angle], 'right_hip_angle': [right_hip_angle]})

    # Predict pose using the loaded model. Adjust this to your model's method if necessary.
    probabilities = loaded_model.predict_proba(features_df)[0]
    max_probability = np.max(probabilities)
    prediction = loaded_model.classes_[np.argmax(probabilities)]
    
    # Check if the maximum probability is below the threshold
    if max_probability < threshold:
        label = "Unknown Pose"
    else:
        label = prediction

    # Choose a color based on the pose classification
    if label != 'Unknown Pose':
        color = (0, 255, 0)  # Green for known poses
    else:
        color = (0, 0, 255)  # Red for unknown poses

    # Overlay the pose label on the image
    cv2.putText(output_image, label, (30, 60), cv2.FONT_HERSHEY_TRIPLEX, 2, color, 2)

    # Display the output image with the pose label, if specified
    if display:
        plt.figure(figsize=[10,10])
        plt.imshow(output_image[:,:,::-1]); plt.title("Pose Classification"); plt.axis('off');
        plt.show()
    else:
        return output_image, label
```

# Image Classifier

```python
def classifyPoseInFile(file_path, pose):
    # Check if the file is an image (you might want to add more image extensions)
    if file_path.endswith(('.jpg', '.jpeg', '.png')):
        # Read the image
        image = cv2.imread(file_path)

        # Perform pose detection
        output_image, landmarks = detectPose(image, pose, display=False)

        # If landmarks are detected, classify the pose
        if landmarks:
            classifyPose(landmarks, output_image, display=True)

# Example usage
file_path = "media/playbook/File_1.jpg"  # Adjust this path to your file
classifyPoseInFile(file_path, pose)
file_path = "media/playbook/File_2.jpg"  # Adjust this path to your file
classifyPoseInFile(file_path, pose)
file_path = "media/playbook/File_3.jpg"  # Adjust this path to your file
classifyPoseInFile(file_path, pose)
```

# Video Classifier

```python
# Video Classification

import cv2
import mediapipe as mp

# Initialize MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, model_complexity=1)

def process_video(video_path):
    # Initialize video capture
    cap = cv2.VideoCapture(video_path)
    
    # Define the codec and create VideoWriter object to save the video
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter('output_video.avi', fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        # Perform pose detection on the frame
        output_image, landmarks = detectPose(frame, pose, display=False)
        
        if landmarks:
            # Classify the pose
            output_image, _ = classifyPose(landmarks, output_image, display=False)
        
        # Write the frame with the classification overlay into the output file
        out.write(output_image)
        
        # Display the frame in real-time
        cv2.imshow('Pose Classification', output_image)
        
        # Check for key press
        key = cv2.waitKey(1)
        
        # Break the loop if 'q' or Esc key is pressed
        if key & 0xFF == ord('q') or key == 27:
            break
    
    # Release resources
    cap.release()
    out.release()
    cv2.destroyAllWindows()

# Example usage
video_path = 'media\playbook\plank.mp4'  # Adjust this path to your video file
process_video(video_path)
```
